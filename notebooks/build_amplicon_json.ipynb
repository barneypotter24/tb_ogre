```python
import pandas as pd
import json
```

```python
df = pd.read_csv("../data/meta/primers_amplicons.tsv", sep="\t")
df
```

```python
d = {
    "name": "MTB Moldova primer scheme",
    "amplicons": []
}

for index, row in df.iterrows():
    start = row["Start"]
    end = row["End"]
    d["amplicons"].append([start,end])
d
```

```python
with open("../rampart/protocols/md2026/primers.json", "w") as f:
    print(json.dumps(d, indent=4), file=f)
```
